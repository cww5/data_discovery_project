## Summarize the data Set

```python
nums = [cons_df.shape[1], gasyearly_df.shape[1], coned_df.shape[1], 
    heatcool_df.shape[1], incs_df.shape[1], #hel1_df.shape[1], hel2_df.shape[1],
   tons_df.shape[1]]
```

```python
sum(nums) - len(nums) + 1
```

```python
all_columns = coned_df.merge(gasyearly_df, on='year', how='inner')
all_columns2 = all_columns.merge(cons_df, on='year', how='right')
all_columns3 = all_columns2.merge(heat_df, left_on='year', right_on='calendar_year', how='left')
all_columns4 = all_columns3.merge(cool_df, left_on='year', right_on='calendar_year', how='left', suffixes=('_heat', '_cool'))
all_columns5 = all_columns4.merge(num_complaints, on='year', how='left')
all_columns6 = all_columns5.merge(tons, on='year', how='left')
all_columns6.sort_values(by='year', inplace=True)
all_columns6.columns
```

```python
all_columns6.shape # should be 40x39
```

```python
all_cols = list(all_columns6.columns)
all_cols.remove('year')
all_cols.insert(0, 'year')
```

```python
final_data = all_columns6.loc[:, all_cols]
final_data.reset_index(inplace=True, drop=True)
final_data.head(10)
```

```python
final_data['calendar_year_change_from_base_year_%_heat'] = final_data['calendar_year_change_from_base_year_%_heat'].astype(np.float64)
final_data['calendar_year_change_from_previous_year_%_heat'] = final_data['calendar_year_change_from_previous_year_%_heat'].astype(np.float64)
final_data['fiscal_year_change_from_base_year_%_heat'] = final_data['fiscal_year_change_from_base_year_%_heat'].astype(np.float64)
final_data['fiscal_year_change_from_previous_year_%_heat'] = final_data['fiscal_year_change_from_previous_year_%_heat'].astype(np.float64)
```

```python
final_data.dtypes
```

```python
from sklearn.preprocessing import minmax_scale
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
num_data = final_data.select_dtypes(include=numerics)
num_data = num_data.drop('year', axis=1)
num_data.head(10)
```

```python
stand_data = pd.DataFrame(minmax_scale(num_data, feature_range=(0, 1), axis=0, copy=True), columns=num_data.columns)
```
