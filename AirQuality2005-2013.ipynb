```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 999)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
air_quality = 'https://data.cityofnewyork.us/resource/c3uy-2p5r.json'
airQ_df = pd.read_json(air_quality)
airQ_df = airQ_df[(airQ_df.geo_type_name == 'Borough')]
```

```python
airQ_df = airQ_df[(airQ_df.year_description == '2005')]
airQ_df.plot(kind='bar',x='geo_entity_name',y='data_valuemessage',color='red')
plt.show()

airQ_df = pd.read_json(air_quality)
airQ_df = airQ_df[(airQ_df.geo_type_name == 'Borough')]
airQ_df.rename(columns={'year_description' : 'year',})
airQ_df = airQ_df[(airQ_df.year_description == '2005')]
airQ_df.plot(kind='bar',x='geo_entity_name',y='data_valuemessage',color='red')
plt.show()
```
